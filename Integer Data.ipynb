```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.plotly as py
import plotly.graph_objs as go
```

```python
path = './df.xlsx'
```

```python
df = pd.read_excel(path)
```

```python
df.head()
```

```python
cleanup = {'Type' : {'Accident':1, 'Incident':0},
           'Amateur Built' : {'Yes':1, 'No':0},
           'Injury Severity' : {'Non-Fatal':0,'Fatal':1}}
df.replace(cleanup, inplace=True)

# for i in df:
#     i['Type'] = i['Type'].map({'Accident':1, 'Incident':0}).astype(int)
#     i['Amateur Built'] = i['Amateur Built'].map({'Yes':1, 'No':0}).astype(int)
#     i['Injury Severity'] = i['Injury Severity'].map({'Non-Fatal':0,'Fatal':1}).astype(int)

df.head()
```

```python
df = pd.get_dummies(df, columns=["Report Status", "Purpose of Flight", "Aircraft Damage", "Aircraft Category"]
                    , prefix=['Report', 'Purpose', 'Damage', 'Aircraft Category'])
df.head()
```

```python
writer = pd.ExcelWriter('./df.xlsx')
df.to_excel(writer,"df")
writer.save()
```
